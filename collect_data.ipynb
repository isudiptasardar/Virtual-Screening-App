```python
import pandas as pd
from chembl_webresource_client.new_client import new_client as searcher
```

# Search For the Target

```python
target_search = searcher.target.search('Aromatase')
searched_df = pd.DataFrame.from_dict(target_search)
searched_df
```

### Select the target
In this case the first one with index [0]

```python
selected_target = searched_df['target_chembl_id'][0]
selected_target
```

## Now download activity data of the selected target from ChEMBL server

```python
downloader = searcher.activity.filter(target_chembl_id = selected_target).filter(standard_type = 'IC50') #Get IC50 only
data = pd.DataFrame.from_dict(downloader)
data
```
